```python
# Importing all necessary libraries
import cv2
import os
  
# Read the video from specified path
cam = cv2.VideoCapture("/content/drive/Shareddrives/AI-Drone/Dataset/Outdoor/0721 Outdoor/Video/Raw Data/Matrice200V2_dp100m.mp4")

try:
    # creating a folder named data
    if not os.path.exists('Moving_Matrice200V2_dp100m'):
        os.makedirs('Moving_Matrice200V2_dp100m')

# if not created then raise error
except OSError:
    print ('Error: Creating directory of data')
  
# frame
currentframe = 16102
  
while(True):
      
    # reading from frame
    ret,frame = cam.read()
  
    if ret:
        # if video is still left continue creating images
        name = './Moving_Matrice200V2_dp100m/Moving_Matrice200V2_dp100m_0721_' + str(currentframe) + '.jpg'
        print ('Creating...' + os.getcwd() + name)
  
        # writing the extracted images
        if currentframe % 90 == 0:
          cv2.imwrite(name, frame)
  
        # increasing counter so that it will
        # show how many frames are created
        currentframe += 1
    else:
        break
  
# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```

```python
import shutil
import os

# Copy Files from dir1 to dir2
dir1 = "/content/drive/Shareddrives/AI-Drone/Dataset/Outdoor/0721 Outdoor/Video/Training Data/Moving_Matrice200V2_dp100m/"
dir2 = "/content/drive/Shareddrives/AI-Drone/Dataset/Outdoor/0721 Outdoor/Video/clf/100m/"

files = os.listdir(dir1) # Get file list from dir1

for file in files:
    shutil.copy(dir1 + file, dir2 + file) # Copy files from dir1 to dir2 using shutil.copy module
    print ('Creating...' + dir2 + file)
```
